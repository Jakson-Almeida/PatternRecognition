Método do holdout

```python
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
clf=MLPClassifier(hidden_layer_sizes=(100),activation='tanh', random_state=10, max_iter=40000).fit(X_train, y_train)
y_pred = clf.predict(X_test)
Perro = 1-accuracy_score(y_test, y_pred)
print("Prob de Erro:", Perro)
print(confusion_matrix(y_test, y_pred))
```

Método leave-one-out

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
erros=np.zeros(len(X))
aux=0
for i in range(len(X)):
    X_train=np.delete(X,i,0)
    y_train=np.delete(y,i)
    X_test=X[i]
    y_test=y[i]
    scaler = MinMaxScaler()
    clf=MLPClassifier(hidden_layer_sizes=(100,),activation='tanh', random_state=10, max_iter=40000).fit(X_train, y_train)
    y_pred = clf.predict(X_test.reshape(1, -1))
    if y_pred!=y_test:
      aux=aux+1;
      erros[i]=i;
```

```python
Perro = aux/len(X)
print("Prob de Erro:", Perro)
```
